```python
#Para pintar y ver el mapa de Calor
def check_nan(df: pd.DataFrame) -> None:
    
    nan_cols  = df.isna().mean()*100
    
    display(f'N nan cols: {len(nan_cols[nan_cols>0])}')
    display(nan_cols[nan_cols>0])
    
    plt.figure(figsize=(10,6))

    sb.heatmap(df.isna(),
               yticklabels=False,
               cmap='viridis',
               cbar = False)

    plt.show();
```

### Date, Year y Time

```python
#Agrupamos las columnas de Date y Year en un mismo grupo al cual denominamos fehcas con el cual sera la que trabajaremos
fechas = data[['Date', 'Year', 'Time']]
fechas.head()
```

```python
def quitar(string):
    
    if len(string) > 11:
        string = string[len(string)-11:len(string)]
    
    return string
```

```python
fechas['Date'] = fechas['Date'].apply(quitar)
fechas['Date']
```

```python
#Para modificar las columnas de Date y Year, para solo poner el año en el que ocurrio el ataque.
fechas['Year'] = fechas['Date'].apply(lambda x : x[len(x)-4:len(x)])
fechas['Year'] = fechas['Year'].apply(lambda x : 'No hay' if x == 'ente' else x)
list(fechas['Year'])
```

```python
for i in range(data.shape[0]):
    
    if data['Case Number.1'][i] == 'Unknown':
        fechas['Time'][i] = 'None'
        
list(fechas['Time'].unique())
```

```python
fechas['Time'] = fechas['Time'].fillna('Unknown')
fechas['Time'].unique()
```

```python
check_nan(fechas)
```
